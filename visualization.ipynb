```
import pandas as pd
from matplotlib import pyplot as plt
```

# 히스토그램
 - Pandas를 이용한 iris(붓꽃) 데이터셋 로드
 - 붓꽃 데이터셋은 4가지 특성: 꽃받침(sepal)과 꽃잎(petal) 각각의 너비, 길이
 - 붓꽃 데이터의 4가지 특성 값의 분포를 히스토그램으로 각각 시각화 



# Pandas 데이터 로드


*   Pandas의 Dataframe 객체에 iris(붓꽃) 데이터셋의 'data' 항목 데이터를 로드
  
  (data, target, target_names, feature_names 등)
*  **get_iris-df( )** 함수는 dataframe(data(iris 4가지 특성), speacies로 구성) 객체 반환

```
#iris dataset
import sklearn.datasets

def get_iris_df():
    ds = sklearn.datasets.load_iris()
    df = pd.DataFrame(ds['data'], columns=ds['feature_names'])
    code_species_map = dict(zip(range(3), ds['target_names']))
    df['species'] = [code_species_map[c] for c in ds['target']]
    return df

df = get_iris_df()
df
```

# 원형 차트
*   데이터셋 분석 중에 사용하거나, 이진 분류 결과 정리할 때 주로 사용
*   



```
sums_by_species = df.groupby('species').sum()
var = 'Sepal width(cm)'
sums_by_species['sepal width (cm)'].plot(kind='pie', fontsize=20)
plt.ylabel(var, horizontalalignment='left')
plt.title(var + 'classification iris', fontsize=25)
plt.show()
```

# 막대 그래프

*   여러 값의 상대적인 크기를 쉽게 비교
*   하단은  붓꽃의 꽃받침 너비로 분류한 부채 붓꽃, 무지개색 붓꽃, 버지니아 붓꽃 차트



```
sums_by_species = df.groupby('species').sum()
var = 'sepal width (cm)'
sums_by_species[var].plot(kind='bar', fontsize=15, rot=30)

plt.title(var + ' classification iris', fontsize=20)
plt.show()
```



*   하단은 붓꽃 종류에 따른 꽃받침 길이/너비, 꼬칲 길이/너비 차트

```
sums_by_species = df.groupby('species').sum()
sums_by_species.plot(kind='bar', subplots=True, fontsize=12)

plt.suptitle('classification by species')
plt.show()
```

# 히스토그램

*   Pandas를 이용한 iris(붓꽃) 데이터셋 로드
*   붓꽃 데이터셋은 4가지 특성: 꽃받침(sepal)과 꽃잎(petal) 각각의 너비, 길이
*   붓꽃 데이터의 4가지 특성 값의 분포를 히스토그램으로 각각 시각화

```
df.plot(kind='hist', subplots=True, layout=(2,2))
plt.suptitle('iris histogram', fontsize=20)
plt.show()
```

```
for spec in df['species'].unique():
    forspec = df[df['species']==spec]
    forspec['petal length (cm)'].plot(kind='hist', alpha=0.4, label=spec)
    
plt.legend(loc='upper right')
plt.suptitle('petal length by species')
plt.savefig('iris_hist_by_spec.png')
```
